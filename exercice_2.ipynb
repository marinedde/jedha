```python
"""Étape 1 : Classe Store
Cette classe de base contient les attributs et méthodes communs aux magasins."""


class Store:
    def __init__(self, name, address, customers=None):
        """Initialise un magasin avec un nom, une adresse, et une liste de clients.
        """
        self.name = name
        self.address = address
        self.customers = customers if customers is not None else []



    def add_customer(self, customer_name):
        """
        Ajoute un client à la liste des clients.
        """
        self.customers.append(customer_name)

    def compute_number_of_customers(self):
        """
        Retourne le nombre total de clients.
        """
        return len(self.customers)

```

```python
"""Étape 2 : Classe GarageStore (hérite de Store)
Ajoute des fonctionnalités spécifiques à un garage, comme les types de voitures.

"""

class GarageStore(Store):

    def __init__(self, name, address, customers=None, car_types=None):

        """

        Initialise un garage avec des types de voitures en plus des attributs de Store.

        """

        super().__init__(name, address, customers)  # Appelle le constructeur de Store

        self.car_types = car_types if car_types is not None else []



    def add_car_type(self, car_type):

        """

        Ajoute un type de voiture au garage.

        """

        self.car_types.append(car_type)



    def get_car_types(self):

        """

        Retourne la liste des types de voitures.

        """

        return self.car_types



"""Explications :

• super() : Permet d’appeler le constructeur de la classe parente (Store).

• Attribut spécifique : car_types pour stocker les types de voitures.

• Méthodes spécifiques : Ajouter et retourner les types de voitures."""



"""Étape 3 : Classe GroceryStore (hérite de Store)

Ajoute des fonctionnalités spécifiques à une épicerie, comme les types de nourriture et le calcul des ventes totales."""



class GroceryStore(Store):

    def __init__(self, name, address, customers=None, food_types=None, sales=None):

        """

        Initialise une épicerie avec des types de nourriture et des ventes.

        """

        super().__init__(name, address, customers)  # Appelle le constructeur de Store

        self.food_types = food_types if food_types is not None else []

        self.sales = sales if sales is not None else []



    def add_food_type(self, food_type):

        """

        Ajoute un type de nourriture à l'épicerie.

        """

        self.food_types.append(food_type)



    def add_sale(self, sale_amount):

        """

        Ajoute une vente à la liste des ventes.

        """

        self.sales.append(sale_amount)



    def compute_total_sales(self):

        """

        Retourne le total des ventes.

        """

        return sum(self.sales)



"""Explications :

• Attributs spécifiques : food_types pour les types de nourriture et sales pour les ventes.

• Méthodes spécifiques : Ajouter des types de nourriture, enregistrer des ventes, et calculer le total des ventes."""



#Étape 4 : Exemple d’utilisation



# Importation des fonctions depuis utils.py si nécessaire

# from utils import ...



# Création d'une instance de Store

store = Store("Generic Store", "123 Main Street")

store.add_customer("Alice")

store.add_customer("Bob")

print(f"Nombre de clients : {store.compute_number_of_customers()}")  # Résultat : 2



# Création d'une instance de GarageStore

garage = GarageStore("AutoFix", "456 Garage Lane")

garage.add_customer("Charlie")

garage.add_car_type("SUV")

garage.add_car_type("Sedan")

print(f"Types de voitures : {garage.get_car_types()}")  # Résultat : ['SUV', 'Sedan']



# Création d'une instance de GroceryStore

grocery = GroceryStore("Fresh Foods", "789 Grocery Blvd")

grocery.add_customer("David")

grocery.add_food_type("Fruits")

grocery.add_food_type("Vegetables")

grocery.add_sale(100)

grocery.add_sale(50)

print(f"Total des ventes : {grocery.compute_total_sales()}")  # Résultat : 150



"""Résumé :

1. Classe Store : Contient les attributs et méthodes communs (nom, adresse, clients, etc.).

2. Classe GarageStore : Hérite de Store et ajoute des fonctionnalités spécifiques aux garages (types de voitures).

3. Classe GroceryStore : Hérite de Store et ajoute des fonctionnalités spécifiques aux épiceries (types de nourriture, total des ventes).

4. Héritage : Permet de réutiliser et d’étendre les fonctionnalités de la classe parent."""

```
